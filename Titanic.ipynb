```python
def fill_age(row):
    Age = row['Age']
    Pclass = row['Pclass']

    if pd.isnull(Age):
        if Pclass == 1:
            return 38.233441  # Replace missing values with 37 for Pclass 1
        elif Pclass == 2:
            return 29.877630  # Replace missing values with 29 for Pclass 2
        else:
            return 25.140620  # Replace missing values with 24 for Pclass 3
    else:
        return Age  # Return the age if it's not missing


df['Age'] = df.apply(fill_age, axis=1)


print(df['Age'].isnull().sum())
```
